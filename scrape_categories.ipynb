```python
import requests
from scraper import UNBISThesaurusScraper
from consts import HEADERS
from bs4 import BeautifulSoup
```

```python
url = "https://metadata.un.org/thesaurus/categories?lang=en"
html = UNBISThesaurusScraper.get_html(url)
```

```python
parser = BeautifulSoup(html, "html.parser")
```

```python
# Find by class name
class_ = "row collapsible"
raw_topics = parser.find_all(class_=class_)
```

```python
len(raw_topics)
```

```python
for raw_topic in raw_topics:
    topic_name = raw_topic.find(class_="bc-link domain").text
    print(topic_name)
    subtopics = raw_topic.find_all(class_="row mx-2")
    for subtopic in subtopics:
        print(f"\t{subtopic.text.strip()}")
```

```python
import json
import random

topics = {
    "nodes": [],
    "edges": [],
}

COLORS = {"topic": "#FA5A3D", "subtopic": "#5A75DB"}

for i, raw_topic in enumerate(raw_topics):
    topic_name = raw_topic.find(class_="bc-link domain").text
    subtopics = raw_topic.find_all(class_="row mx-2")
    color = "#"+''.join([random.choice('0123456789ABCDEF') for i in range(6)])
    topic_json = {
        "key": topic_name.split(" - ")[0],
        "attributes": {
            "type_": "topic",
            "label": topic_name.strip(),
            "color": COLORS["topic"]
        }
    }
    topics["nodes"].append(topic_json)
    for subtopic in subtopics:
        subtopic_name = subtopic.text.strip()
        subtopic_json = dict(
            key=subtopic_name.split(" - ")[0],
            attributes={
                "type_": "subtopic",
                "label": subtopic.text.strip(),
                "color": COLORS["subtopic"]
            }
        )
        subtopic_edge = dict(
            key=f"{topic_name.split(' - ')[0]}->{subtopic_name.split(' - ')[0]}",
            source=topic_name.split(" - ")[0],
            target=subtopic_name.split(" - ")[0],
            attributes={}
        )
        topics["nodes"].append(subtopic_json)
        topics["edges"].append(subtopic_edge)
        


with open("../../../un-graph-ui/data.json", "w") as f:
    json.dump(topics, f)
```
